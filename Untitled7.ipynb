```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# -------------------------------
# Step 1 : Load Dataset (SAFE WAY)
# -------------------------------

file_path = "C:\\Users\\Dharshikhan V.R\\Downloads\\stock_market_risk_dataset.csv"

# Check if file exists
if not os.path.exists(file_path):
    print("❌ File not found. Check the path or file name.")
    print("\nFiles in Downloads folder:")
    print(os.listdir(r"C:\Users\ARUL KUMARAN.T\Downloads"))
    exit()

# Load dataset
data = pd.read_csv(file_path)

# -------------------------------
# Step 2 : Overview
# -------------------------------
data.columns = data.columns.str.strip()

print("Dataset Preview:")
print(data.head())

print("\nDataset Info:")
print(data.info())

# -------------------------------
# Step 3 : Select Numeric Data
# -------------------------------
numeric_data = data.select_dtypes(include=np.number)

if numeric_data.shape[1] == 0:
    print("❌ No numeric columns found")
    exit()

prices = numeric_data.iloc[:, 0].to_numpy()

# -------------------------------
# Step 4 : Basic Statistics
# -------------------------------
print("\nAverage Price:", np.mean(prices))
print("Minimum Price:", np.min(prices))
print("Maximum Price:", np.max(prices))

# -------------------------------
# Step 5 : Returns
# -------------------------------
returns = np.diff(prices) / prices[:-1]

print("\nReturns:")
print(returns)

# -------------------------------
# Step 6 : Risk
# -------------------------------
risk = np.std(returns)
print("\nRisk (Std Dev):", risk)

# -------------------------------
# Step 7 : Moving Average
# -------------------------------
window = 3
moving_avg = np.convolve(prices, np.ones(window)/window, mode='valid')

# -------------------------------
# Plot 1 : Line Plot
# -------------------------------
plt.figure()
plt.plot(prices, marker="o")
plt.title("Stock Price Trend")
plt.xlabel("Time")
plt.ylabel("Price")
plt.grid(True)
plt.show()

# -------------------------------
# Plot 2 : Bar Plot
# -------------------------------
plt.figure()
plt.bar(range(len(prices)), prices)
plt.title("Stock Prices")
plt.xlabel("Time")
plt.ylabel("Price")
plt.show()

# -------------------------------
# Plot 3 : Histogram
# -------------------------------
plt.figure()
plt.hist(returns, bins=10)
plt.title("Returns Distribution")
plt.xlabel("Returns")
plt.ylabel("Frequency")
plt.show()

# -------------------------------
# Plot 4 : Moving Average
# -------------------------------
plt.figure()
plt.plot(prices, label="Price")
plt.plot(range(window-1, len(prices)), moving_avg, label="Moving Avg")
plt.legend()
plt.title("Moving Average")
plt.show()

# -------------------------------
# Plot 5 : Heatmap (if multiple columns)
# -------------------------------
if numeric_data.shape[1] > 1:
    plt.figure()
    sns.heatmap(numeric_data.corr(), annot=True)
    plt.title("Correlation Heatmap")
    plt.show()
```
